```python
import matplotlib.pyplot as plt

file_types = ['HTML', 'Images','JavaScript', 'CSS', 'Fonts', 'Plain Text', 'JSON']
url_counts = [43, 198, 190, 34, 31, 29, 22] 

plt.figure(figsize=(10, 6))
bars = plt.bar(file_types, url_counts)
plt.title('Number of URLs in WARC File by File Type')
plt.xlabel('File Type')
plt.ylabel('Number of URLs')
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height() 
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, 
             str(yval), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
